```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/Data/customer.csv")
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[["Poor", "Average", "Good"], ["School", "UG", "PG"]])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
oe.categories_
```

## Label Encoding

- This is for output columns only.
- This transform should be used to encode target values, i.e y and not the input X.

- It randomly decide which class will get which value, we can not select that.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
