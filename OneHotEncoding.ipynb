- We generally don't use pd.get_dummies while model creation because pandas does not remember what value he gave to which category and it may change when we run another time. And we have to transform in same way both the train and test data after learning from train data. Which is not possible with this pd.get_dummies.

- That's why for this we use sklearn OneHotEncoder class.

# 4. OneHotEncoding with Top Categories

- Brand category which has more than 100 count will be taken and rest all will be clubed into one category.

```python
counts = df["brand"].value_counts()
```

```python
threshold = 100 # you can take yours
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, "uncommon"))
```
